```python
mng_db = client['cice_db']          # Replace mongo db name
collection_name = 'tips_all'       # Replace mongo db collection name
mng_db.drop_collection(collection_name)

db_cm = mng_db[collection_name]
    
data = pd.read_csv('./datos/tips.csv', index_col = ['sex', 'day'])
data = data.sort_index()
#La colecci贸n va a tener un mont贸n e resguistros, un registro para fila de la tabla
jj = data.to_json(orient='split')
d_j = json.loads(jj)
db_cm.insert_one(d_j)

```

```python
# guardo el json
with open('tipsall.json', 'w') as f:
     json.dump(jj, f)
        
config = json.loads(open('tipsall.json').read())
config
```

### Crear una base de datos y una colecci贸n 3露

```python
data = pd.read_csv('./datos/tips.csv')

tabla_year = pd.DataFrame({'year': [2014, 2015, 2014, 2015],
                        'place': ['El recuerdo', 'El olvido', 'El olvido', 'La Leyenda'],
                        'estadisticas' : [data.iloc[0:30], data.iloc[31: 80],
                                         data.iloc[81:200], data.iloc[201:]]  })


mng_db = client['cice_db']          # Replace mongo db name
collection_name = 'tips_years'       # Replace mongo db collection name

# borro y creo la colecci贸n
mng_db.drop_collection(collection_name)
db_cm = mng_db[collection_name]
jj = tabla_year.to_json(orient='records')
d_j = json.loads(jj)
db_cm.insert_many(d_j)
```

```python

    
with open('tipsyears.json', 'w') as f:
     json.dump(jj, f)
        
config = json.loads(open('tipsyears.json').read())
type(config)
```
