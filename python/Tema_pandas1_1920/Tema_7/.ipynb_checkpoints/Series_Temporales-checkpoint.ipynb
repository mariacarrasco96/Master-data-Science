## Acceso a los datos

Se puede extraer los datos asociados a un rango de fechas de manera muy natural. Por ejemplo, para acceder a los datos de los años `2005` y  `2007` escribimos lo siguiente:

```python
dax['2005':'2007']
```

Para rangos que impliquen los meses o días, escribimos:

```python
dax['2005-06':'2006-06-15']
```

También podemos recuperar una fecha concreta:

```python
dax['2015-06']
```

## Resampling

La función `resample` permite agrupar datos por años `A`, meses `M` o días `D`. Dicha función devuelve grupos de tuplas, sobre los cuales se puede realizar alguna de las funciones de agregación, como `mean()`, `std()`, `min()`, etc. 

Por ejemplo, para calcular el valor medio de cada una de las columnas por meses, escribimos lo siguiente:

```python
dax.resample('M').mean()
```

Para calcular el valor máximo __en periodos de 3 años__ para  cada una de las columnas, escribimos lo siguiente:

```python
dax.resample('3A').max()
```

También podemos aplicar varias funciones de agregación a cada una de las columnas. Por ejemplo, queremos el máximo y mínimo __cada 18 meses__ para la columna `Close` y la media cada 18 mesese de la columna `Adj Close`:

```python
d = {'Close': [max, min], 'Adj Close': np.mean}
dax.resample('18M').apply(d)
```

## Truncate

El método  `truncate` de la clase DataFrame permite seleccionar solo los datos comprendidos entre dos valores `before`y `after`. Por ejemplo, para seleccionar los valores comprendicos entre '2017-08-15' y  '2017-09-01', escribimos:

```python
dax.truncate(before = '2017-08-15', after = '2017-09-01')
```

## Operación rolling

Pandas permite el estudio de tendencias mediante la aplicación de funciones específicas para el cálculo de medias móviles. El cálculo de la media móvil se realiza mediante la operación  `rolling` de las Series junto con la función `mean()`. También se puede aplicar otras funciones como `min()`, `std()`, etc.

Por ejemplo, podemos calcular la media móvil del valor `Close` a 42 días y a 252 días. 
Creamos dos nuevas columnas:

```python
dax['media 252d'] = dax.Close.rolling(window = 252, min_periods = 10).mean()
dax.head(13)
```

Calcula una nueva columna llamada `media 42d` para almacenar la media móvil a 42 días. 


```python
dax['media 42d'] = dax.Close.rolling(window = 42, min_periods = 10).mean()
dax.head(13)
```
