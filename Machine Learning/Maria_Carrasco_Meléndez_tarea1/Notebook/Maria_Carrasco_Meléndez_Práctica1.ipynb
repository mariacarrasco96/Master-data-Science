## Introducción

El siguiente dataset se encuentra centrado en la financiación, deuda, inflación y crisis sistemática, la cual tuvo lugar entre 1860 y 2014. Se han recogido los datos de 13 países africanos como Egipto, Algeria o Angola entre otros.



## Objetivo del trabajo

Con objeto de comprender con más detenimiento cuales fueron las causas de esta crisis ocurrida en África entre 1860 y 2014, este documento estudia las variables que influyeron en este acontecimiento estudiando que relaciones hay entre ellas.

Variables empleadas para comprender la dinámica de la estabilidad financiera dentro del contexto africano:

* __"CC3"__: Código del pais.
* __"Systematic_crisis"__: Con el valor de 0 decimos que no hubo crisis en el año, y al contrario con el 1.
* __"Exch_usd"__: Tipo de cambio del país frente al USD.
* __"domestic_debt_in_default"__: Con el valor de 0 no incumplimiento de la deuda interna soberana, al contrario con el 1.
* __"sovereign_external_debt_default"__: Con el valor de 0 no incumplimiento de la deuda externa soberana, al contrario con el 1.
* __"gdp_weighted_default"__: Deuda total en mora frende al PIB.
* __"inflation_annual_cpi"__: Tasa de inflación IPC anual.
* __"independence"__: Con el valor de 0 indicamos la no independencia, al contrario con el 1.
* __"currency_crisis"__: Con el valor de 0 no ocurrió crisis monetaria, al contrario con el valor de 1.
* __"inflation_crisis"__: Con el valor de 0 no ocurrió ninguna crisis de inflación en el año, al contrario con el valor de 1.
* __"banking_crisis"__: Disponemos de dos valores: "no_crisis" y "crisis".

Al final del siguiente documento , se le concederá una importancia especial a la inflación debido a las importantes consecuencias que provoca esta última sobre la economía. Por este motivo, se formulará un modelo predictivo que trate de explicar esta última en función de las demás variables.

## 4. Dummies

Transformamos en dummies las variables categóricas y lo concatenamos con nuestro dataframe crisis_3, creando uno nuevo llamado crisis_new donde tan solo se han conservado las variables numéricas

```python
crisis_3=crisis.drop(['cc3','case','gdp_weighted_default','year'],axis=1)
crisis_3
```

```python

dummies = pd.get_dummies(crisis_3)
dummies = dummies.drop(['exch_usd', 'inflation_annual_cpi'], axis = 1)
crisis_3 = crisis_3.drop(['country', 'systemic_crisis', 'domestic_debt_in_default', 'sovereign_external_debt_default', 'independence','currency_crises','inflation_crises','banking_crisis'], axis = 1)
crisis_new = pd.concat([crisis_3, dummies], axis=1)
```

```python
crisis_new
```

## 5. Train y test

Se prepara la librería y se crea una divisón donde la variable a predecir será la inflación y el resto de variables son las explicativas de esta última.
Para ello se aprte de un 70% de train y un 30% de test y se ven estas divisiones.

```python

from sklearn.model_selection import train_test_split



y = crisis_new.iloc[:, 6] #Variable que queremos explicar ("inflacion")
x = crisis_new.loc[:, crisis_new.columns != "inflacion"] #Variables que la explican (las demás)


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=123)


print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

Vemos la dimensión de la partición

## 6. Modelo Lasso

```python
from sklearn import linear_model
from sklearn.linear_model import LassoCV
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

## Formulamos el modelo

lassocv = LassoCV(alphas = None, cv = 10, max_iter = 100000, normalize = True)
lassocv.fit(x_train, y_train)
lassocv.alpha_



model_lasso = Lasso(alpha=lassocv.alpha_) # funcion lasso
model_lasso.fit(x_train, y_train) # ajuste sobre el train
pred_lasso = model_lasso.predict(x_test) # prediccion sobre el test
print(np.sqrt(mean_squared_error(y_test,pred_lasso))) # error sobre el test
error_lasso = np.sqrt(mean_squared_error(y_test,pred_lasso))
print(r2_score(y_test, pred_lasso)) # R^2 sobre el test
```

El error de este modelo es ínfimo.

## 7. Conclusión

Concluimos que el tipo de cambio y la inflación no estan correlacionadas por lo que no son influyentes la una sobre la otra, sin embargo podemos predecir con mucha exactitud la inflación en relación al resto de las variables.
